#  Relationship between Suicide Rates/100K and GDP per Capita



```python
import seaborn as sns
def reg_plot_suicide_rates_vs_gdp():
    '''
    This is the dataframe we need to make our regression plot. 
    
    '''
    df = suicide_data.groupby(["country", "year"])[['suicides/100k pop']].sum().unstack()['suicides/100k pop']
    df.columns.name = df.index.name = None
    df_new = suicide_data.groupby(["country", "year"])[['gdp_per_capita ($)']].min().unstack()['gdp_per_capita ($)']
    df_new.columns.name = df_new.index.name = None
    df = pd.concat([df,df_new])
    df = df[(df.index == 'United States') | (df.index == 'Luxembourg') | 
                                       (df.index == 'Singapore') | (df.index == 'Switzerland') |(df.index == 'Lithuania')]
    
    brazil = df[df.index == 'Singapore'].T
    brazil.columns = ['Suicides/100k', 'GDP per Capita']
    
    US =  df[df.index == 'United States'].T
    US.columns = ['Suicides/100k', 'GDP per Capita']
    
    luxem  = df[df.index == 'Luxembourg'].T
    luxem.columns = ['Suicides/100k', 'GDP per Capita']
    
    para  = df[df.index == 'Switzerland'].T
    para.columns = ['Suicides/100k', 'GDP per Capita']
    
    lithu  = df[df.index == 'Lithuania'].T
    lithu.columns = ['Suicides/100k', 'GDP per Capita']
    
    return pd.concat([brazil, US, luxem, para])


all_countries = reg_plot_suicide_rates_vs_gdp()
all_countries

sns.regplot(x="Suicides/100k", y="GDP per Capita", data = all_countries)
plt.title('Relationship between Suicide Rates/100K and GDP per Capita')

```
